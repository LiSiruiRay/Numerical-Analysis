## Lemma 1:
$\forall a,b \in \mathbb{R}^{+} \cup \{0\}$, $a \ge b \iff a^2 \ge b^2$

## Proof of Lemma 1:
($\Rightarrow$) We are given $a,b \in \mathbb{R}^{+} \cup \{0\}$ and $a \ge b$. We have $a,b \ge 0$. If we multiply both sides by $a$, we get:
$$a * a \ge b * a$$
Now, since $a \ge b$, we can replace $a$ with $b$ in the right side of the equation without affecting the relationship:
$$a * a = a^2 \ge b^2 = b * b $$

($\Leftarrow$) We are given $a,b \in \mathbb{R}^{+} \cup \{0\}$ and $a^2 \ge b^2$. We have $a,b \ge 0$

Assume $a < b$, 

&emsp;&emsp; 1. If $b \ne 0$, $a \cdot a < b \cdot a < b \cdot b$, which implies $a^2 < b^2$, $\Rightarrow \Leftarrow$


&emsp;&emsp; 2. If $b = 0$, then $a < 0$, $\Rightarrow \Leftarrow$
    
Therefore, if $a^2 \ge b^2$, $a \ge b$.

```python
from pylab import *
import math

def dft(theta,n):
    f_values = cos(n * sin(theta/2) * pi)
    dft = rfft(f_values)
    modulus = absolute(dft)
    return modulus

n=128
theta=arange(0, 2*pi,2*pi/n)
wavenumber = rfftfreq(n, 2*pi/n)


plot(wavenumber, dft(theta,2), 'b-', 
     wavenumber, dft(theta,4), 'g-', 
     wavenumber, dft(theta,8),  'r-', 
     wavenumber, dft(theta,16), 'y-', )
xlabel("Wavenumber", fontsize=20)
ylabel("|C\u2096|", fontsize=20)
legend(["n = 2", "n = 4", "n = 8", "n = 16"],
        loc="lower left")
yscale('log')   
xscale('log')
```

```python
from pylab import *
import math

def dftDerivative(f_values,d_theta):
    dft = rfft(f_values)
    k = rfftfreq(len(f_values), d_theta)
    dft_derivative = dft * (1j * 2 * pi * k)
    derivative = irfft(dft_derivative, len(f_values))
    return derivative

def f(theta):
    return cos(4 * sin(theta/2))

def f_p(theta):
    return -2 * sin(4 * sin(theta/2)) * cos(theta/2)

interval = [2**n for n in range(3,10)]
absolute_errors = []
for n in interval:
    theta = arange(0, 2*pi,2*pi/n)
    d_theta = 2*pi/n
    f_values = f(theta)
    fp_values = f_p(theta)
    derivative = dftDerivative(f_values, d_theta)
    max_error = max(fabs(fp_values - derivative))
    absolute_errors.append(max_error)

    
    figure()
    plot(theta, derivative, label='DFT Derivative')
    title(f'Derivative of f(theta) with n={n} points')
    xlabel('Theta')
    ylabel('Derivative')
    legend()
    show()

plot(interval, absolute_errors)
xlabel("Number of Points", fontsize=20)
ylabel("Absolute Error", fontsize=20)
title("Absolute Error")

yscale('log')   
xscale('log')
```

```python
from pylab import *
import math

def dftDerivative(f_values,d_theta):
    dft = rfft(f_values)
    k = rfftfreq(len(f_values), d_theta)
    dft_derivative = dft * (1j * 2 * pi * k)
    derivative = irfft(dft_derivative, len(f_values))
    return derivative

def f(theta):
    return 1/(1+pow((theta - pi),2))

def f_p(theta):
    return -2 * (theta-pi)/pow((pow((theta-pi),2) +1),2)

interval = [2**n for n in range(3,10)]
absolute_errors = []
for n in interval:
    theta = arange(0, 2*pi,2*pi/n)
    d_theta = 2*pi/n
    f_values = f(theta)
    fp_values = f_p(theta)
    derivative = dftDerivative(f_values, d_theta)
    max_error = max(fabs(fp_values - derivative))
    absolute_errors.append(max_error)

    
    figure()
    plot(theta, derivative, label='DFT Derivative')
    title(f'Derivative of f(theta) with n={n} points')
    xlabel('Theta')
    ylabel('Derivative')
    legend()
    show()

plot(interval, absolute_errors)
xlabel("Number of Points", fontsize=20)
ylabel("Absolute Error", fontsize=20)
title("Absolute Error")

yscale('log')   
xscale('log')
```
